```python
import pandas as pd
from sklearn import preprocessing
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
import seaborn as sns
```

```python
# Read songs dataset with labels and tags
data_frame = pd.read_csv("final_dataset.csv",)
data_frame
```

```python
# Mood identification and recommendation
cluster_info = []
mood_info = []
for i in range(3):
    value1 = input("Please enter song name:\n") 
    found1 = data_frame[data_frame['song_name'].str.contains(value1, na=False, case=False)]
    cluster_info.append(found1['label'].tolist()[0])
    mood_info.append(found1['tag'].tolist()[0])

# over here we identify the mood based on user inputs   
cluster = max(cluster_info,key=cluster_info.count)
mood = max(mood_info,key=mood_info.count)


data_frame_temp = data_frame[ data_frame['label']==cluster ]
data_frame_temp = data_frame_temp[ data_frame['tag']==mood ]


data_frame_temp.drop_duplicates(subset ="song_name", keep = False, inplace = True) 

# over here we recommend the list of songs.
data_frame_temp = data_frame_temp.sort_values(by=['song_popularity'],ascending = False)[0:50].sample(10)

data_frame_temp.sort_values(by=['song_popularity'],ascending = False)
```
